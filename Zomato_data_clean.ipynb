<h2>Deleting Unwanted Columns</h2>

```python
df.columns
```

----

<p><i> Divide the columns to keep & drop</p>

```python
col_to_keep=['name', 'online_order', 'book_table', 'rate', 'dish_liked','approx_cost(for two people)']

col_to_drop=['url', 'address','phone','votes', 'location', 'rest_type','cuisines','reviews_list', 'menu_item',
       'listed_in(type)', 'listed_in(city)']
```

```python
df.shape
```

```python
len(col_to_keep)
```

```python
len(col_to_drop)
```

```python
# modified df
df=df[col_to_keep]
```

```python
df
```

```python
type(df['name'])  # ----> series
```

```python
type(df[['name']])  # ----> dataframe
```

---------

<h2>Renaming Columns</h2>

```python
new_col_names=[]
for i in df.columns:
    new_col_names.append(i.upper())
```

```python
new_col_names
```

```python
df.columns=new_col_names
df
```

<h2>Drop Duplicates</h2>

```python
df.duplicated().value_counts()
```

```python
# check the no. of duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

<h2>Removing NaN values</h2>

```python
df.isna().sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

----

<h2>Reset Index</h2>

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns='index', inplace=True)
```

```python
df
```

------

<h2>Formatting Index</h2>

```python
df['RATE'].info()
```

```python
# remove the blank space
df['RATE']=df['RATE'].str.replace(" ","")
```

```python
df['RATE']=df['RATE'].str.replace('/5','')
```

------

<h2>Boolean values to Numeric</h2>

```python
df['ONLINE_ORDER']=df['ONLINE_ORDER'].apply(lambda x: 1 if x=="Yes" else 0)
```

```python
df['BOOK_TABLE']=df['BOOK_TABLE'].apply(lambda x: 1 if x=="Yes" else 0)
```

-----

<h2>Export</h2>

```python
df.to_csv("Zomato_data_clean.csv",index=False)
```

```python
df.to_excel("Zomato_data_clean.xlsx",index=False)
```
